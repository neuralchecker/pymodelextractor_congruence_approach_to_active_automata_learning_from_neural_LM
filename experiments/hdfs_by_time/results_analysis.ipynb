```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load data from experiment

```python
evaluation_data = pd.read_csv('./results/results_05_03_2023_23_40_54.csv')
```

```python
len(evaluation_data)
```

```python
evaluation_data.head(16).sort_values('Number of Extracted States')
```

```python
evaluation_data['Total_Acc'] = evaluation_data['Accuracy']*(1-evaluation_data['Unknown Results Percentage']) 
```

```python
evaluation_data.head(16).sort_values('Number of Extracted States')
```

```python
evaluation_data['Error'] = 1 - evaluation_data['Total_Acc']
```

```python
evaluation_data
```

```python
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
p1 = sns.lineplot(data=evaluation_data, x ="TimeBound", y = "Number of Extracted States", marker = "o", label = "|Q|")
p1.grid(False)
p2 = sns.lineplot(data=evaluation_data, x ="TimeBound", y = "EquivalenceQuery", marker = "D", label = "EQ")
p2.grid(False)
ax2 = plt.twinx()
p3 = sns.lineplot(data=evaluation_data, x="TimeBound", y="Error", color = 'g', ax=ax2, marker = "s", label = "Error")
p3.grid(False)
p1.set_xlabel("Time Bound", fontsize = 12)
p1.set_ylabel("Amount (|Q| and EQ)", fontsize = 12)
p3.set_ylabel("Percentage (Error)", fontsize = 12)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
sns.set_style("darkgrid")
div = 1.6
a4_dims = (11.7/div, 8.27/div)
fig = plt.figure(figsize=a4_dims)
#fig = plt.figure()
ax = fig.add_subplot(111)

lns1 = ax.plot(evaluation_data['TimeBound'], evaluation_data["Number of Extracted States"], '-s', label = "|Q|")
lns2 = ax.plot(evaluation_data['TimeBound'], evaluation_data["EquivalenceQuery"], '-o', label = 'EQ', color = 'g')
ax2 = ax.twinx()
lns3 = ax2.plot(evaluation_data['TimeBound'], evaluation_data["Error"], '-D', label = 'Error', color = 'r')

# added these three lines
lns = lns1+lns2+lns3
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc="upper center")
ax.grid()
ax.set_xlabel("Time Bound (seconds)")
ax.set_ylabel("Amount (|Q| and EQ)")
ax2.set_ylabel("Percentage (Error)")
ax.grid(False)
ax2.grid(False)
sns.move_legend(
    ax, "lower center",
    bbox_to_anchor=(.5, 1), ncol=3, title=None, frameon=False,
)
plt.show()
```
